## Exercise 1

A coffee shop is having a sale on coffee and tea. 

On day 1, 29 bags of coffee and 41 bags of tea were sold, for a total of 490 dollars.

On day 2, they sold 23 bags of coffee and 41 bags of tea, for which customers paid a total of 448 dollars.  

How much does each bag cost?

```python
# Create and solve the relevant system of equations
import numpy as np

A = np.array([[29, 41], [23, 41]])
b = np.array([490, 448])
x = np.linalg.solve(A, b)

print("The price of one bag of coffee is", x[0], "dollars.")
print("The price of one bag of tea is", x[1], "dollars.")
```

# Describe your result

The result of the calculation using NumPy is that each bag of coffee costs 7 dollars and each bag of tea costs 9 dollars. This means that the coffee shop is selling coffee at a price of 7 dollars per bag and tea at a price of 9 dollars per bag. We arrived at this result by using matrix algebra and solving a system of linear equations, which was represented in matrix form.

## Exercise 2

The cost of admission to a popular music concert was 162 dollars for 12 children and 3 adults. 

The admission was 122 dollars for 8 children and 3 adults in the same music concert. 

How much was the admission for each child and adult?

```python
# Create and solve the relevant system of equations
import numpy as np

A = np.array([[12, 3], [8, 3]])
b = np.array([162, 122])
x = np.linalg.solve(A, b)

print("The cost of admission for each child is", x[0], "dollars.")
print("The cost of admission for each adult is", x[1], "dollars.")
```

# Describe your result
The result of the calculation using NumPy is that the cost of admission for each child is 10 dollars and the cost of admission for each adult is 36 dollars. This means that the music concert charged 10 dollars per child and 36 dollars per adult for admission. We arrived at this result by using matrix algebra and solving a system of linear equations, which was represented in matrix form.

## Exercise 3

You want to make a soup containing tomatoes, carrots, and onions.

Suppose you don't know the exact mix to put in, but you know there are 7 individual pieces of vegetables, and there are twice as many tomatoes as onions, and that the 7 pieces of vegetables cost 5.25 USD in total. 
You also know that onions cost 0.5 USD each, tomatoes cost 0.75 USD and carrots cost 1.25 USD each.

Create a system of equations to find out exactly how many of each of the vegetables are in your soup.

```python
# Create and solve the relevant system of equations
import numpy as np

# coefficients of the variables
A = np.array([[0.75, 1.25, 0.5], [0, 1, 3]])

# constants on the right-hand side of the equations
b = np.array([5.25, 7])

# solve for onion
o = (b[1] - A[1,1]*2) / (A[1,2])

# solve for carrot
c = (b[0] - A[0,0]*2*o - A[0,2]*o) / (A[0,1])

# solve for tomato
t = 2 * o

# print the solution
print("There are", round(t), "tomatoes,")
print(round(c), "carrots, and")
print(round(o), "onions in the soup.")
```

## Exercise 4

A landlord owns 3 properties: a 1-bedroom, a 2-bedroom, and a 3-bedroom house. 

The total rent he receives is 1240 USD. 

He needs to make some repairs, where those repairs cost 10% of the 1-bedroom house’s rent. The 2-bedroom repairs cost 20% of the 2-bedroom rental price and 30% of the 3-bedroom house's rent for its repairs.  The total repair bill for all three houses was 276 USD. 

The 3-bedroom house's rent is twice the 1-bedroom house’s rent. 

How much is the individual rent for three houses?

```python
# Create and solve the relevant system of equations
import numpy as np

A = np.array([[1, 1, 1], [0.1, 0.2, 0.3], [2, 0, -1]])
b = np.array([1240, 276, 0])
try:
    x = np.linalg.solve(A, b)
    print("The rent for the 1-bedroom house is:", round(x[0]), "USD")
    print("The rent for the 2-bedroom house is:", round(x[1]), "USD")
    print("The rent for the 3-bedroom house is:", round(x[2]), "USD")
except np.linalg.LinAlgError:
    print("There is no consistent solution to this system of equations.")

```
